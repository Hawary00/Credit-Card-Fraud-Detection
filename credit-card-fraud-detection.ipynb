Plotting the Correlation Matrix:

The correlation matrix graphically gives us an idea of how features correlate with each other and can help us predict what are the features that are most relevant for the prediction.

```python
# dividing the X and the Y from the dataset
X = df.drop(['Class'], axis = 1)
y = df['Class']
print(X.shape)
print(y.shape)
```

convert to numpy array with no columns

```python
xData = X.values
yData = y.values
```

```python
xData.shape
```

```python
yData.shape
```

Taraining and test data spliting

```python
from sklearn.model_selection import train_test_split
xTrain, xTest, yTrain, ytest = train_test_split(xData, yData, test_size = 0.2, random_state = 42)
```

```python
xTrain.shape
```

```python
yTrain.shape
```

> Bulding Random forest model using scikit learn

```python
from sklearn.ensemble import RandomForestClassifier
RFC = RandomForestClassifier()
RFC.fit(xTrain, yTrain)
```

```python
yPred = RFC.predict(xTest)
```

```python
n_outliers = len(fraud)
# print(n_outliers)
n_errors = (yPred != yTest).sum()
print("The model used is Random Forest classifier it has number of errors ",n_errors,"from total ", n_outliers)
```

Building all kinds of evaluating parameters


```python
from sklearn.metrics import classification_report, accuracy_score
from sklearn.metrics import precision_score, recall_score

acc = accuracy_score(yTest, yPred)
print("The accuracy is {}".format(acc))

perc = precision_score(yTest, yPred)
print("The percision is {}".format(perc))

rec = recall_score(yTest, yPred)
print("The recall is {}".format(rec))
```

```python
CR = classification_report(yTest, yPred)
print("classification_report{}".format(CR))
```

```python
from sklearn.metrics import confusion_matrix

LABLES = ['Normal', 'Fraud']
conf_matrix = confusion_matrix(yTest, yPred)
plt.figure(figsize=(12, 10))
sns.heatmap(conf_matrix, xticklabels = LABLES, yticklabels = LABLES, annot = True, fmt ="d")
plt.title("Confusion matrix")
plt.ylabel('True class')
plt.xlabel('Predicted class')
plt.show()
```
